وهناك ثلاثة اصناف لحالات الطوارئ وهي EMS وتخص الحالات الصحية ثم بعدها نوع الحالة مثلا اغماء او كسر في العظام والصنف الثاني هو الحريق Fire وبعدها سبب الحريق مثلا بسبب الغاز او اي شيء اخر والصنف الثالث هو المرور Traffic ويأتي بعده نوع المشكلة مثلا اصدام سيارة

EMS: BACK PAINS/INJURY في العمود الجديد ستكون فقط الجزء الاول من هذه القيمة وهو EMS

ومن اجل ذلك نحتاج الى صناعة وظيفة def بأستخدام بايثون تقسم عمود العنوان title بأستخدام النقطتين : والتي تفصل بين الجزء الاول والجزء الثاني من العنوان وتقوم الوظيفة بأرجاع فقط الجزء الاول من العنوان
<br>
ثم نقوم بعمل لوب بأستخدام for او while من اجل تنفيذ هذه الوظيفة على جميع قيم العمود title حتى تفصل جزئين كل قيمة في عمود العنوان
<br><br>
هذا الشيء ممكن ولكن سيستغرق وقت كبير جدا خاصة في حالة كانت عدد قيم العمود (عدد سطور العمود) كبيرة جدا لانه سنصنع وظيفة وسنقوم بتنفيذ هذه الوظيفة على كل قيمة في العمود بأستخدام لوب وربما تستغرق هذه العملية الكثير من الوقت والجهد
<br>
ولكن هناك حل بسيط وكفوء لهذه المشكلة حيث لغة بايثون توفر لنا طرق اخرى لأنشاء وظيفة غير def حيث بأمكاننا انشاء وظيفة بأستخدام lambda
<br><br>
lambda هي احدى طرق انشاء وظيفة بلغة بايثون وهي طريقة مختصرة وسريعة وكفوئة حيث تأخذ مدخل واحد مثلا x ثم نقطتين بعد اسم المدخل ثم العملية التي نريد ان تؤديها هذه الوظيفة مثلا قسم جزئين القيمة x بأستخدام (:) وارجاع فقط الجزء الاول الذي تحدثنا عنه سابقا والاندكس للجزء الاول اكيد صفر والجزء الثاني اندكسه واحد ولكننا نريد فقط الجزء الاول
<br><br>

وبعد انشاء وظيفة بطريقة مباشرة بأستخدام lambda نريد الان تطبيق هذه الوظيفة على جميع قيم العمود title ونستطيع القيام بذلك من خلال لوب ولكن كما ذكرنا سابقا اللوب يستهلك وقت ولحسن الحظ يوجد لدينا مثود رائعة اسمها apply حيث بأمكاننا استخدام هذه المثود على العمود title وكتابة داخل القوسين الوظيفة التي نريد تنفيذها على كل قيمة في العمود title
<br><br>
يبدو ان الشرح طويل ولكنها في الحقيقة سهلة جدا ومهمة جدا حيث ستستخدمها في مسيرتك المهنية كثيرا حيث في الكود الذي في الاسفل ببساطة سوف ننشئ عمود جديد سوف نسميه السبب Reason وقيمته فقط الجزء الاول من عمود title وسوف نقوم بذلك من خلال تنفيذ وظيفة تأخذ كل قيمة في العمود title وتسميها x ثم تقسم هذه x بأستخدام المثود split على اساس النقطتين ثم ترجع فقط الجزء الاول والذي اندكسه صفر

</p>


```python
df['Reason'] = df['title'].apply(lambda x: x.split(':')[0])
```

الان سوف ننشئ عمود جديد قيمته هي تحويل عمود timeStamp الى تاريخ ووقت بأستخدام المثود to_datetime من مكتبة Pandas

</p>

```python
df["timeStamp"] = pd.to_datetime(df["timeStamp"] )
df["timeStamp"].head(3)
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>
وكما هو واضح عمود التاريخ والوقت يحتوي على التاريخ سنة وشهر ويوم ويحتوي على الوقت ساعة ودقيقة وثانية

<br>
الان سوف ننشئ ثلاث اعمدة جديدة وهي عمود الساعة ويمكننا استخراجه من عمود timeStamp بأستخدام المثود hour ولذلك نحتاج الى صناعة وظيفة بأستخدام lambda تستخرج فقط الساعة من عمود التاريخ والوقت timeStamp ثم ننفذ هذه الوظيفة على كل قيمة في هذا العمود بأستخدام apply كما تعلمنا سابقا

<br>
وبنفس الطريقة سنصنع عمود ثاني اسمه الشهر ولكنا بأستخدام المثود month


</p>

```python
df["hour"] = df['timeStamp'].apply(lambda x: x.hour)
df["month"] = df['timeStamp'].apply(lambda x: x.month)
df["month"].head(3)
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>
والعمود الثالث الذي سوف نصنعه هو عمود يوم الاسبوع وبأمكاننا استخراجه بأستخدام المثود dayofweek ولكن المشكلة هي انه سنحصل على قيم من 0 الى 6 كل قيمة تمثل يوم في الاسبوع مثلا القيمة صفر هي يوم الاثنين والواحد هي الثلاثاء والخ..
<br>
نحن نحتاج الى تحويل كل قيمة الى اليوم الذي تمثله من اجل جعلها مفهومة للأشخاص العاديين ومن اجل ذلك سوف نصنع دكشنري مفاتيحه هي رقم اليوم وقيمته هي اليوم الذي يمثله هذا الرقم

</p>

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>
والان سوف نصنع وظيفة تطبق هذا الدكشنري على كل قيمة في عمود التاريخ والوقت وترجع لنا بأسم اليوم وليس رقم اليوم كما في السابق وسوف نخزنها في عمود جديد اسمه يوم الاسبوع

</p>

```python
df["Day_of_week"] = df['timeStamp'].apply(lambda x: dmap[x.dayofweek])
df["Day_of_week"].head(3)
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

حيث سنستخدم الرسم countplot لاننا هنا ايضا مهتمين بالمقارنة بين الاعداد مثلا مقارنة عدد الحرائق في يوم الاثنين مع باقي الايام وستكون ال x لهذا الرسم هي عمود يوم الاسبوع وستكون hue هي عمود السبب والمقصود في ال hue هنا هي اعمدة قيم x حيث سيكون لنا في كل قيمة من x عدد من الاعمدة وهذه الاعمدة هي hue

<br> 
حيث x هي ايام الاسبوع و hue هي السبب وهي ثلاث اعمدة

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

الان نريد استكشاف ماهي الايام وماهي الساعات التي تكثر فيها مكالمات الطوارئ وهذا الشيء مهم جدا ومن المؤكد سيطلب منك تحليله واستكشافه في هكذا نوع من البيانات وفي انواع البيانات المشابهة مثلا بيانات خدمة العملاء لشركة معينة
<br><br>
لأستكشاف ذلك نحتاج الى رسم يكون فيه محور ال X هو ساعات اليوم ومحور ال Y هو ايام الاسبوع
<br>
لذلك سوف نصنع DataFrame جديدة تكون عبارة عن تجميع العمودين عمود ايام الاسبوع وعمود الساعات من خلال استخدام الوظيفة groupby وطريقة التجميع هي العدد وسوف نسمي هذا الجدول dayhour1

<br>

حيث سيكون الاندكس لهذه ال DataFrame هي عمود ايام الاسبوع وعمود الساعة مثلا الاندكس الاول يوم الجمعة وفيه 24 اندكس هي ساعات هذا اليوم وكذلك الحال لباقي السطور 
</p>

```python
dayHour1 = df.groupby(by=['Day_of_week','hour']).count()
dayHour1
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>
الان سنقوم بشيء جديد ومفيد جدا وهو فك اندكس هذه ال DataFrame الجديدة حيث سنجعل الاندكس لهذا الجدول هي ايام الاسبوع والاعمدة هي الساعات من خلال فكها بأستخدام المثود unstack وقيم هذا الجدول هي قيم عمود السبب
<br>
وللتوضيح اكثر حيث جدولنا الجديد dayHour1 يحتوي على اندكس ايام الاسبوع والساعات وبأستخدام unstack سيتم تفكيكه ونشره حيث سيتحول كل يوم من ايام الاسبوع الى سطر وكل ساعة من ساعات اليوم الى عمود وهذا الجدول الجديد يحتاج قيم هذه القيم ستكون عمود السبب Reason من خلال استخدام unstack عليها وسوف نسمي الجدول الجديد dayHour2

</p>

```python
dayHour2 = dayHour1["Reason"].unstack()
dayHour2
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

</p>
